# **Q-4 Write a Python program to swap two variables.**

```python
a = 4
b = 5
print(f'this is original no. a-> {a} b-> {b}')

temp = a
a = b
b = temp

print(f'this is swaped no. a-> b :{a}')
print(f'this is swaped no. b-> a :{b}')
```
