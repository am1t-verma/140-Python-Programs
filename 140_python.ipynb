# **Q-4 Write a Python program to swap two variables.**

```python
a = 4
b = 5
print(f'this is original no. a-> {a} b-> {b}')

temp = a
a = b
b = temp

print(f'this is swaped no. a-> b :{a}')
print(f'this is swaped no. b-> a :{b}')
```

# **Q-6 Write a Python program to convert kilometers to miles**

```python
# Conversion factor: 1 kilometer = 0.621371 miles

kilo = float(input('Enter the value to convert Kill -> Miles'))
miles = kilo * .621371
print(f'{kilo} kilometers is equal to : {miles}')
```

# **Q-7 Write a Python program to convert Celsius to Fahrenheit**

```python
# Conversion formula: Fahrenheit = (Celsius * 9/5) + 32

celsius = float(input('enter a value to convert into Celsius'))

fahre_1 = (celsius * 9/5) + 32

print(f'{celsius} celsius is : {fahre_1} Fahrenheit')
```

# **Q-9 Write a Python Program to Check if a Number is Positive, Negative or Zero.**

```python
no = float(input('enter a no.'))

if no == 0:
    print(f'given no. : {no} is Zero')

elif no > 0 :
    print(f'given no. : {no} is Positive')

else:
    print(f'given no. : {no} is Negative')
```
